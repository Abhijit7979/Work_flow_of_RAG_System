#### Loading documents using langchain

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader("../3.2-dataIngestion/speech.txt")
docs=loader.load()
docs
```

#### converting into chunks using langchain

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_doc_chunks=text_splitter.split_documents(docs)
print(final_doc_chunks[0:3]) # first 3 chunks 
```

#### vector Embedding and vector StoreDB

```python
from langchain_community.vectorstores import Chroma 
# chroma db is a open source vector database 
db=Chroma.from_documents(final_doc_chunks,embedding_dimension,collection_name="openai_embedding")
db
```

Retrieve the results from query vectorstore db by applying similarity search.

```python
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query,k=1)
print(retrieved_results)
```
