#### Sentence Transformers on Hugging Face 

hugging face sentence-transformers is a python framework for state of the art sentence, text and image embeddings. One of the embedding models is used in the HuggingFaceEmbeddings class.We have also added an alias for SentenceTransformerEmbeddins for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="This is a test document of hugging face embeddings."
query_result=embedding.embed_query(text)
query_result
```

```python
len(query_result) # dimension of the embedding is 384 
```

```python
doc_result=embedding.embed_documents([text,"hello! my name is abhijit and i am a Gen Ai engineer. "])

# Check the dimensions of the document embeddings

len(doc_result[0]),len(doc_result[1])
```
